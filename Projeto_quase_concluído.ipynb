```python
from tabulate import tabulate
opcao = ""
matricula = []
nome = []
mediaPonderada = []
while opcao != "SAIR": 
    opcao = input("Bem vindo ao sistema de cadastro \n Selecione uma das opções a seguir: \n A - CADASTRAR UM ALUNO \n B - CONSTULTAR UM CADASTRO \n C - APAGAR UM CADASTRO \n D - SOLICITAR O RELATÓRIO DOS CADASTROS QUE ESTÃO NA BASE \n SAIR - FECHAR O SISTEMA DE CADASTRO \n")
    while opcao == "A": 
        print("#### TELA DE CADASTRO ####")
        matAluno = input("Digite a matrícula do aluno: \n")
        matricula.append(matAluno)
        nomeAluno = input("Digite o nome do aluno: \n")
        nome.append(nomeAluno)
        mediaPond = float(input ("Digite a média desse aluno: \n"))
        mediaPonderada.append(mediaPond)
        validacao = "E"
        while validacao == "E":
            validacao = input("Você gostária de realizar mais um cadastro? S ou N\n")
            if validacao == "S":
                opcao = "A"
                validacao = ""
            elif validacao == "N":
                opcao = ""
                validacao = ""
            else: 
                print("Opção inexistente, tente novamente")
                validacao = "E"
    while opcao == "B":
        print("### TELA DE CONSULTA ###")
        pesquisa = input("Digite o código, nome ou média do aluno:\n")
        naoEncontrado = ""
        for x in matricula:
            if x == pesquisa: 
                posicao = matricula.index(pesquisa)
                print("####Código encontrado no sistema, segue as informações do estudante####")
                print("\nCódigo: "+str(pesquisa)+"\nNome: "+str(nome[posicao])+"\nMédia: "+str(mediaPonderada[posicao]))  
        for x in nome: 
            if x == pesquisa: 
                posicao = nome.index(pesquisa)
                print("####Nome encontrado no sistema, segue as informações do estudante####")
                print("\nCódigo: "+str(matricula[posicao])+"\nNome: "+str(nome[posicao])+"\nMédia: "+str(mediaPonderada[posicao]))
        validacaoEntrada = pesquisa.isnumeric()
        if validacaoEntrada == True:
            conversaoPesquisa = float(pesquisa)
            for x in mediaPonderada: 
                if x == conversaoPesquisa: 
                    posicao = mediaPonderada.index(conversaoPesquisa)
                    print("#### Média encontrada no sistema, segue as informações do estudante ####")
                    print("\nCódigo: "+str(matricula[posicao])+"\nNome: "+str(nome[posicao])+"\nMédia: "+str(mediaPonderada[posicao]))
        validacao = "E"
        while validacao == "E":
            validacao = input("Você gostária de realizar mais uma pesquisa? S ou N\n")
            if validacao == "S":
                opcao = "B"
                validacao = ""
                pesquisa = ""
            elif validacao == "N":
                opcao = ""
                validacao = ""
            else: 
                print("Opção inexistente, tente novamente")
                validacao = "E"  
    while opcao == "C":
        print("#### TELA DE EXCLUSÃO ####")
        exclusao = input("Digite o nome ou o código para realizara exclusão\n")
        for x in matricula: 
            if x == exclusao: 
                posicao = matricula.index(exclusao)
                validador = input("Código encontrado, gostária de porsseguir com a exclusão? (DIGITE S PARA CONTINUAR)\n")
                if validador == "S":
                    del matricula[posicao]
                    del nome[posicao]
                    del mediaPonderada[posicao]
                    print("Informações excluídas com sucesso")
                else:
                    print("Operação cancelada")
        for x in nome:
            if x == exclusao: 
                posicao = nome.index(exclusao)
                validador = input("Nome encontrado, gostária de porsseguir com a exclusão? (DIGITE S PARA CONTINUAR)\n")
                if validador == "S":
                    del nome[posicao]
                    del matricula[posicao]
                    del mediaPonderada[posicao]
                    print("Código excluído com sucesso")
                else:
                    print("Operação cancelada")
        validacao = "E"
        while validacao == "E":
            validacao = input("Você gostária de realizar mais uma exclusão? S ou N\n")
            if validacao == "S":
                opcao = "C"
                validacao = ""
            elif validacao == "N":
                opcao = ""
                validacao = ""
            else: 
                print("Opção inexistente, tente novamente")
                validacao = "E"
    while opcao == "D":
        print("### Relatório com todas as informações da BASE DE DADOS ### ")
        listRelatorio = [] 
        for x in matricula:
            listOrganizada = []
            posicao = matricula.index(x)
            listOrganizada.append(matricula[posicao])
            listOrganizada.append(nome[posicao])
            listOrganizada.append(mediaPonderada[posicao])
            listRelatorio.append(listOrganizada)
        print(tabulate(listRelatorio, headers=["Matricula","Aluno","Média"]))
        validacao = "E"
        while validacao == "E":
            validacao = input("Você gostária gerar mais um relatório? S ou N\n")
            if validacao == "S":
                opcao = "D"
                validacao = ""
            elif validacao == "N":
                opcao = ""
                validacao = ""
            else: 
                print("Opção inexistente, tente novamente")
                validacao = "E"
```
